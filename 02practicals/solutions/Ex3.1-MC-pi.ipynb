## Ex. 3.1: Ordinary MC Integration

```python
import random
N = 1_000_000

count = 0  #count will store the number of random points
           #that fell within the 1st quadrant of the unit circle

for i in range(N):
    x, y = random.random(), random.random()
    if x**2 + y**2 < 1:
        count += 1
    
print(4*count/N)
```
